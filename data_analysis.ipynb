# Loading Data

```python
import json

def load_json(file_name):
    """
    Loads and returns data from a JSON file.

    Args:
        file_name (str): Path to the JSON file.

    Returns:
        dict or list: Parsed JSON content.
    """
    with open(file_name, 'r') as file:
        data = json.load(file)
    return data

# #example usage
# data = load_json('./files/clicked_points_20250324_112218.json')
# data
```

```python
data_dict_1     = load_json('./files/dict_clicked_points_20250630_124212.json')
# data_dict_2     = load_json('./files/dict_250320_depth_and_size_vs_density.json')
# data_dict_3     = load_json('./files/dict_250220_depth_and_size_vs_density.json')

main_dict = {}
main_dict.update(data_dict_1)
# main_dict.update(data_dict_2)
# main_dict.update(data_dict_3)
```

# Extracting the data

```python
import numpy as np

list_of_all_depths_new             = []
list_of_all_areas_new              = []
list_of_all_angles_new             = []
list_of_all_minors_over_majors_new = []
list_of_all_distances_new          = []

for scan_name in main_dict.keys():
    for hole_number in range(main_dict[scan_name]['total_number_of_clicked_points']):
#        print(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['depth'])
        list_of_all_depths_new.append(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['depth'])

        area = np.pi * main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['major_axis'] * main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['minor_axis']
#        print(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['major_axis'])
        list_of_all_areas_new.append(area)
        list_of_all_angles_new.append(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['theta'])
        list_of_all_minors_over_majors_new.append(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['minor_over_major'])
        list_of_all_distances_new.append(main_dict[scan_name]['hole_number_{hole_number}'.format(hole_number=hole_number+1)]['distance_h_e'])
```

# Plotting the data
